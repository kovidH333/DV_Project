# **PREPROCESSING**

```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv("nba2k-full.csv")
data
```

```python
data['jersey'] = data['jersey'].str.replace('#', '').astype(int)
data['jersey']
```

```python
data['salary'] = data['salary'].str.replace('$', '').astype(int)
data['salary']
```

```python
data['height'] = data['height'].str.split('/').str[1].astype(float)
data['height']
```

```python
data['weight'] = data['weight'].str.split('/').str[1]
data['weight']
```

```python
data['weight'] = data['weight'].str.replace('kg.','').astype(float)
data['weight']
```

```python
import datetime

def calculate_age(birth_date):
    birth_date_obj = datetime.datetime.strptime(birth_date, '%m/%d/%y')
    
    current_date = datetime.date.today()
    
    age = current_date.year - birth_date_obj.year - ((current_date.month, current_date.day) < (birth_date_obj.month, birth_date_obj.day))
    
    return age
```

```python
data['age'] = data['b_day'].apply(calculate_age)
data['age']
```
